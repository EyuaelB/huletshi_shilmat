```python
import pandas as pd
```

```python
df = pd.read_csv("/content/shilmat_data2.csv")
df['Date'] = pd.to_datetime(df.Date)
df=df.drop(columns=['Day'],axis=1) #drop the day column a it is there for recording purpose only , the date is sufficient
df['DayOfTheWeek'] = df.Date.dt.day_of_week
df['DayName'] = df.Date.dt.day_name()
df['Month'] = df.Date.dt.month
df['MonthName'] =df.Date.dt.month_name()
df['DayOfTheMonth'] = df.Date.dt.day
df['Year'] = df.Date.dt.year
```

```python
features = ['DayOfTheWeek','DayOfTheMonth','Month']
X = df[features].values
y = df['Income']
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from lazypredict.Supervised import LazyRegressor

multi_model_tryout = LazyRegressor(ignore_warnings = True ,predictions=True)
model,predictions = multi_model_tryout.fit(X_train,X_test,y_train,y_test)
```
